```python
import os
import sys
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Paths, input, and output file names

outer_dir = os.path.abspath(os.path.join(os.getcwd() ,"../.."))
path = os.path.join(outer_dir, 'IFT_files', 'IFT Results')

stream_list = ['Indian', 'Sproul']
hydro_list = ['unimp','mpof','mmpof','ncifp','tessman']
version = 'V1_0_0'
in_csv = os.path.join(path, 'All_hydrographs_{}.csv'.format(version))
out_csv = os.path.join(path, 'Water_year_types_{}.csv'.format(version))
```

```python
# Import data
df = pd.read_csv(in_csv)
print(len(df))
df.head()
```

```python
# check for missing values
df[df.isna().any(axis=1)]
```

```python
df1 = df.loc[df.ift == 'unimp'].copy()
df2 = (df1.groupby(['stream', 'water_year'])
           .agg({'flow':'mean', 'date':'count'})
           .reset_index()
           .rename(columns={'flow':'wy_mean_flow_cfs', 'date':'day_count'})
          )
# Filter out partial years
df3 = df2.loc[df2.day_count >= 360].copy()
df3
```

```python
df3['pct_rank'] = df3.groupby('stream')['wy_mean_flow_cfs'].rank(pct=True)
df3
```

```python
# Export data to CSV

df3.to_csv(out_csv, encoding='utf-8', index=False)
```
